```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.random.seed(42)
n = 200
data = pd.DataFrame({
    "experience":np.random.randint(0,15,n),
    "education_year": np.random.randint(10,20,n),
    "age": np.random.randint(22,60,n),
    })
data["salary"] = (
    30000 +
    data["experience"] *4000 +
    data["education_year"] * 2000 +
    data["age"] * 500 +
    np.random.randint(n) * 10000
)
data.head()
```

```python
print(data.shape)
print(data.info())
print(data.describe())
(200, 4)
```

```python
corr = data.corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
features = ["experience", "education_year", "age"]

for col in features:
  plt.figure()
  plt.scatter(data[col], data["salary"])
  plt.xlabel(col)
  plt.ylabel("salary")
  plt.show()
```

```python
x = data[["experience", "education_year", "age"]]
y = data["salary"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
    )
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```
